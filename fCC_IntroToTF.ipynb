```python
tensor1 = tf.ones([1, 2, 3])
tensor1
```

```python
tensor2 = tf.reshape(tensor1, [2, 3, 1]) # 2 outside lists, 3 lists inside, 1 item inside
tensor2
```

```python
tensor3 = tf.reshape(tensor2, [3, -1]) # -1 tells the tensor to calculate the size of the dimension in that place
# so in this case it will be [3, 2]
tensor3
```
